```python
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt
```

**Multinomial Distribution**

Multinomial distribution is a generalization of binomial distribution.

It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two. e.g. Blood type of a population, dice roll outcome.

It has three parameters:

**n** - number of times to run the experiment.

**pvals** - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).

**size** - The shape of the returned array.

```python
n = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], size=1000)
print(n)
```

**Explaination**
n=6

Ye number of trials hai. Matlab hum 6 baar experiment karenge.

Example: 6 baar dice roll karna.

pvals=[1/6, 1/6, ..., 1/6]

Ye har outcome ki probability hai.

Dice ke 6 sides ka equal probability hota hai → 1/6 each.

(Sum of all probabilities = 1 hona chahiye ✅).

**Output**
Ye ek array return karega jo batata hai ke 6 trials me har category kitni dafa aayi.

Category 1 (side 1) → 0 times aayi

Category 2 (side 2) → 1 time

Category 3 (side 3) → 2 times

Category 4 (side 4) → 1 time

Category 5 (side 5) → 0 times

Category 6 (side 6) → 2 times

Aur sum(array) = n (6) hamesha hoga.
