```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
```

**Normal Distribution**

It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.

It fits the probability distribution of many events, eg. IQ Scores, Heartbeat etc.

Use the random.normal() method to get a Normal Data Distribution.

It has three parameters:

**loc** - (Mean) where the peak of the bell exists.

**scale** - (Standard Deviation) how flat the graph distribution should be.

**size** - The shape of the returned array.

```python
x = np.random.normal(size=(2, 3))
print(x)
```

```python
x = np.random.normal(loc=5, scale = 1, size=(2, 3))
print(x)
```

```python
a = np.random.normal(size=1000)
print(a)
seaborn.displot(a, kind='kde')
plt.show()
```

**The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.**

**loc=50** → mean 50 hai (data center me 50 ke aas paas hoga)

**scale=5** → spread 5 ka hai (zyada tar values 45–55 ke range me hongi)

**size=1000** → 1000 random numbers generate honge

```python
data = np.random.normal(loc=50, scale=5, size=1000)

print(data[:10])  # first 10 values print karo

# Histogram plot karte hain
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

# seaborn.displot(data, kind='hist')
# seaborn.displot(data, kind='ecdf')

plt.show()
```

```python
# 3 different normal distributions
data1 = np.random.normal(loc=0, scale=1, size=1000)    # mean=0, std=1
data2 = np.random.normal(loc=50, scale=5, size=1000)   # mean=50, std=5
data3 = np.random.normal(loc=100, scale=10, size=1000) # mean=100, std=10

# Plot sab ko ek graph me
plt.figure(figsize=(10,6))

plt.hist(data1, bins=30, alpha=0.5, label="Mean=0, Std=1")
plt.hist(data2, bins=30, alpha=0.5, label="Mean=50, Std=5")
plt.hist(data3, bins=30, alpha=0.5, label="Mean=100, Std=10")

plt.legend()
plt.title("Comparison of Normal Distributions")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
numbers = np.random.normal(loc=100, scale=10, size=1000)
mean = np.mean(numbers)
std_dev = np.std(numbers)

print("Expected mean is 100, Calculated mean:", mean)
print("Expected standard deviation is 10, Calculated standard deviation:", std_dev)
```
